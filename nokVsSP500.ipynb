```python
import os; import pandas as pd; import yfinance as yf; import matplotlib.pyplot as plt

import numpy as np

nok_df = pd.read_csv("nok.csv")
nok_all_df = nok_df.drop(0)


spy_df = pd.read_csv("spy.csv")
spy_all_df = spy_df.drop(0)


print(nok_all_df.head())

nok_returns = nok_all_df['Return'].to_numpy()
spy_returns = spy_all_df['Return'].to_numpy()

nPlusNok = 0
nMinusNok = 0
nPlusSpy = 0
nMinusSpy = 0
twoStdAwayNok = 0
twoStdAwaySpy = 0

for x in nok_returns:
    if x > 0:
        nPlusNok += 1
    elif x < 0:
        nMinusNok += 1
for x in spy_returns:
    if x > 0:
        nPlusSpy += 1
    if x < 0:
        nMinusSpy += 1

for x in spy_returns:
    if abs(x-spy_returns.mean()) > spy_returns.std()*2:
        twoStdAwaySpy +=1
for x in nok_returns:
    if abs(x-nok_returns.mean()) > nok_returns.std()*2:
        twoStdAwayNok += 1

print("Nokia N+: " + str(nPlusNok))
print("Nokia N-: " + str(nMinusNok))
print("Nokia N+: " + str(nPlusSpy))
print("Nokia N-: " + str(nMinusSpy))
print("twoStdNok: " + str(twoStdAwayNok))
print("twoStdS: " + str(twoStdAwaySpy))

x = np.polyfit(nok_returns, spy_returns, 1)

print(x)
```
